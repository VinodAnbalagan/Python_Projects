## Binary Search

```python
def binary_search(arr, item):
  low = 0
  high = len(arr)-1
  while low <= high:
    mid = (low+high)//2
    guess = arr[mid]

    if guess == item:
      return mid

    elif guess > item:
      high = mid - 1

    else:
      low = mid + 1
  return None
```

```python
print(binary_search([1,2,3,4,5], 5)) # Prints the index of iteam in the array
```

## Selection Sort

```python
def findSmallest(arr):
  smallest = arr[0]
  smallest_index = 0
  for i in range (1, len(arr)):
    if arr[i] < smallest:
      smallest = arr[i]
      smallest_index = i
  return smallest_index

def SelectionSort(arr):
  newArr= []
  copiedArr = arr.copy()
  for i in range(len(copiedArr)):
    smallest = findSmallest(copiedArr)
    newArr.append(copiedArr.pop(smallest))
  return newArr

```

```python
print(SelectionSort([5, 3, 6, 2, 10]))
```

## Quicksort

```python
def quicksort(arr):
  if len(arr) < 2:
    return arr
  else:
    pivot = arr[0]
    less = [i for i in arr[1:] if i <= pivot]
    greater = [i for i in arr[1:] if i> pivot]
    return quicksort(less) + [pivot]+ quicksort(greater)

quicksort([1,20,30,2,4,55,21,6])
```

## Bubble Sort

```python
def bubble_sort(nums):
  swapping = True
  end = len(nums)
  while swapping:
    swapping = False
    for i in range(1, end):
      if nums[i-1] > nums [i]:
        nums[i-1], nums[i] = nums[i], nums[i-1]
        swapping = True
    end -= 1
  return nums

bubble_sort([5, 3, 8, 1])
```

## Insertion Sort

```python
def insertion_sort(nums):
    """
    Sorts a list of numbers using the Insertion Sort algorithm.

    Args:
        nums: The list of numbers to sort.

    Returns:
        The same list, sorted in ascending order (Insertion Sort sorts in-place).
    """
    # Insertion sort works by building a sorted portion of the list
    # from left to right.

    # The outer loop iterates through the list, starting from the second element.
    # We assume the first element (at index 0) is already a sorted list of size 1.
    for i in range(1, len(nums)):
        # 'current_element' is the element we want to insert into the
        # already sorted portion of the list (elements from index 0 to i-1).
        current_element = nums[i]

        # 'j' is a pointer that starts at the last element of the sorted portion.
        # We will use 'j' to find the correct position for 'current_element'.
        j = i - 1

        # Inner loop (or shifting loop):
        # While 'j' is not less than 0 (meaning we haven't reached the beginning of the list)
        # AND the element at index 'j' in the sorted portion is greater than
        # the 'current_element' we want to insert...
        while j >= 0 and nums[j] > current_element:
            # ...shift the element at index 'j' one position to the right
            # to make space for the 'current_element'.
            nums[j + 1] = nums[j]
            # Move the pointer 'j' one step to the left to compare with the next element
            # in the sorted portion.
            j -= 1

        # After the inner loop finishes, the correct position for 'current_element'
        # has been found (it's right after the element at index 'j').
        # Place the 'current_element' into its sorted position.
        nums[j + 1] = current_element

        # Optional: Print the list after each pass of the outer loop to see the sorting process
        # print(f"After pass {i}: {nums}")


    # The outer loop finishes when all elements have been processed.
    # Insertion sort sorts the list in-place, so we return the modified original list.
    return nums

# --- Simple Example ---

# The list we want to sort
my_list = [12, 11, 13, 5, 6]

# Call the insertion_sort function on the list
print("Original list:", my_list)
sorted_list = insertion_sort(my_list)

# Print the sorted list
print("Sorted list:", sorted_list) # Note: The original list 'my_list' is now sorted

# --- Another Example ---
my_list_2 = [5, 1, 4, 2, 8]
print("\nOriginal list 2:", my_list_2)
insertion_sort(my_list_2) # Sorts in-place
print("Sorted list 2:", my_list_2)
```

## Merge Sort

```python
def merge(left, right):
    """
    Merges two sorted lists.
    """
    merged = []
    i, j = 0, 0 # Pointers for left and right lists

    # Compare elements and append the smaller one
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1

    # Append remaining elements (at most one list will have them)
    merged.extend(left[i:])
    merged.extend(right[j:])

    return merged

def merge_sort(nums):
    """
    Sorts a list using Merge Sort (minimal version).
    """
    # Base case: list with 0 or 1 element is sorted
    if len(nums) <= 1:
        return nums

    # Divide: find midpoint and split
    mid = len(nums) // 2
    left = nums[:mid]
    right = nums[mid:]

    # Conquer: recursively sort halves and merge
    # The recursive calls return sorted lists, which are then merged
    return merge(merge_sort(left), merge_sort(right))

# --- Simple Example ---

my_list = [12, 11, 13, 5, 6, 7]
sorted_list = merge_sort(my_list)
print("Original list:", my_list)
print("Sorted list:", sorted_list)

# --- Another Example ---
my_list_2 = [38, 27, 43, 3, 9, 82, 10]
sorted_list_2 = merge_sort(my_list_2)
print("\nOriginal list 2:", my_list_2)
print("Sorted list 2:", sorted_list_2)
```

## Queue

```python
from collections import deque
class Queue:
  def __init__(self):
    self.items = deque()

  def push(self, item):
    self.items.append(item)

  def pop(self):
    # Check if the deque is empty
    if len(self.items) == 0:
      return None
    # Pop from the left side (front of the queue) - O(1)
    # This is the optimized part compared to del self.items[0]
    return self.items.popleft()

  def size(self):
    return len(self.items)

  def peek(self):
    # Peeking at the front is still efficient - O(1)
    if len(self.items) == 0:
      return None
    return self.items[0] # Access the first element

# --- Test Cases ---

print("--- Testing Queue ---")

# Test Case 1: Basic FIFO Operations
print("\nTest Case 1: Basic FIFO Operations")
queue1 = Queue()
print(f"Initial size: {queue1.size()}") # Expected: 0
print(f"Peek on empty queue: {queue1.peek()}") # Expected: None
print(f"Pop on empty queue: {queue1.pop()}")   # Expected: None

queue1.push(10) # Add 10
print(f"After push 10: size={queue1.size()}, peek={queue1.peek()}") # Expected: size=1, peek=10

queue1.push(20) # Add 20
print(f"After push 20: size={queue1.size()}, peek={queue1.peek()}") # Expected: size=2, peek=10

queue1.push(30) # Add 30
print(f"After push 30: size={queue1.size()}, peek={queue1.peek()}") # Expected: size=3, peek=10

popped_val = queue1.pop() # Pop (should be 10)
print(f"After pop: popped={popped_val}, size={queue1.size()}, peek={queue1.peek()}") # Expected: popped=10, size=2, peek=20

popped_val = queue1.pop() # Pop (should be 20)
print(f"After pop: popped={popped_val}, size={queue1.size()}, peek={queue1.peek()}") # Expected: popped=20, size=1, peek=30

popped_val = queue1.pop() # Pop (should be 30)
print(f"After pop: popped={popped_val}, size={queue1.size()}, peek={queue1.peek()}") # Expected: popped=30, size=0, peek=None

popped_val = queue1.pop() # Pop from empty
print(f"After pop from empty: popped={popped_val}, size={queue1.size()}, peek={queue1.peek()}") # Expected: popped=None, size=0, peek=None


# Test Case 2: Mixing Operations and Different Types
print("\nTest Case 2: Mixing Operations and Different Types")
queue2 = Queue()
print(f"Initial size: {queue2.size()}") # Expected: 0

queue2.push("apple") # Push string
print(f"After push 'apple': size={queue2.size()}, peek={queue2.peek()}") # Expected: size=1, peek='apple'

popped_val = queue2.pop() # Pop string
print(f"After pop: popped={popped_val}, size={queue2.size()}, peek={queue2.peek()}") # Expected: popped='apple', size=0, peek=None

queue2.push(True) # Push boolean
print(f"After push True: size={queue2.size()}, peek={queue2.peek()}") # Expected: size=1, peek=True

queue2.push(5.5) # Push float
print(f"After push 5.5: size={queue2.size()}, peek={queue2.peek()}") # Expected: size=2, peek=True

print(f"Peek again: {queue2.peek()}") # Expected: True (should not remove)
print(f"Size again: {queue2.size()}") # Expected: 2 (should not change)

popped_val = queue2.pop() # Pop (should be True)
print(f"After pop: popped={popped_val}, size={queue2.size()}, peek={queue2.peek()}") # Expected: popped=True, size=1, peek=5.5

popped_val = queue2.pop() # Pop (should be 5.5)
print(f"After pop: popped={popped_val}, size={queue2.size()}, peek={queue2.peek()}") # Expected: popped=5.5, size=0, peek=None

popped_val = queue2.pop() # Pop from empty
print(f"After pop from empty: popped={popped_val}, size={queue2.size()}, peek={popped_val}") # Expected: popped=None, size=0, peek=None


print("\n--- Testing Complete ---")
```

## Linked List

```python
# Define a Node class to represent each element in the linked list
class Node:
    """
    Represents a single node in a singly linked list.
    """
    def __init__(self, data):
        # The data stored in the node
        self.data = data
        # A pointer (reference) to the next node in the list, initially None
        self.next = None

# Define the LinkedList class
class LinkedList:
    """
    Represents a singly linked list.
    """
    def __init__(self):
        # The head of the list, initially None when the list is empty
        self.head = None

    def append(self, data):
        """
        Adds a new node with the given data to the end of the list.
        This is an O(n) operation in a singly linked list because we
        might need to traverse to the end.
        """
        new_node = Node(data)

        # If the list is empty, the new node becomes the head
        if self.head is None:
            self.head = new_node
            return # Exit the method

        # Otherwise, traverse the list to find the last node
        last_node = self.head
        while last_node.next:
            last_node = last_node.next

        # Link the last node to the new node
        last_node.next = new_node

    def prepend(self, data):
        """
        Adds a new node with the given data to the beginning of the list.
        This is an O(1) operation as we only need to update the head.
        """
        new_node = Node(data)
        # The new node's next pointer should point to the current head
        new_node.next = self.head
        # The new node becomes the new head of the list
        self.head = new_node

    def delete_node(self, key):
        """
        Deletes the first node found with the given data (key).
        This is an O(n) operation in the worst case (node is at the end or not found).
        """
        current_node = self.head

        # Case 1: The head node itself holds the key to be deleted
        if current_node is not None and current_node.data == key:
            self.head = current_node.next # Move the head to the next node
            current_node = None # Dereference the old head node
            return # Exit the method

        # Case 2: The key is not in the head node, search for it elsewhere
        prev_node = None # Keep track of the previous node
        while current_node is not None and current_node.data != key:
            # Traverse the list: move both pointers forward
            prev_node = current_node
            current_node = current_node.next

        # If the key was not found in the list
        if current_node is None:
            print(f"Node with data '{key}' not found in the list.")
            return # Exit the method

        # If the key was found (current_node is not None and current_node.data == key)
        # Unlink the node from the linked list
        prev_node.next = current_node.next

        # Dereference the node to be deleted
        current_node = None

    def display(self):
        """
        Prints the data of each node in the list from head to tail.
        This is an O(n) operation as it requires traversing the entire list.
        """
        current_node = self.head
        elements = []
        while current_node:
            elements.append(str(current_node.data))
            current_node = current_node.next
        print(" -> ".join(elements))

    def search(self, key):
        """
        Searches for a node with the given data (key).
        This is an O(n) operation in the worst case.
        """
        current_node = self.head
        while current_node:
            if current_node.data == key:
                return True # Found the key
            current_node = current_node.next
        return False # Key not found

# --- Simple Test Case ---

print("--- Testing Singly Linked List ---")

# Create an empty linked list
my_list = LinkedList()
print("\nInitial list:")
my_list.display() # Expected: (empty line)

# Test append operation
print("\nAppending nodes:")
my_list.append(1)
my_list.append(2)
my_list.append(3)
my_list.display() # Expected: 1 -> 2 -> 3

# Test prepend operation
print("\nPrepending nodes:")
my_list.prepend(0)
my_list.prepend(-1)
my_list.display() # Expected: -1 -> 0 -> 1 -> 2 -> 3

# Test search operation
print("\nTesting search:")
print(f"Is 2 in the list? {my_list.search(2)}")   # Expected: True
print(f"Is 5 in the list? {my_list.search(5)}")   # Expected: False
print(f"Is -1 in the list? {my_list.search(-1)}") # Expected: True

# Test delete operation
print("\nDeleting nodes:")
my_list.delete_node(2) # Delete node with data 2
print("After deleting 2:")
my_list.display() # Expected: -1 -> 0 -> 1 -> 3

my_list.delete_node(-1) # Delete the head node
print("After deleting -1:")
my_list.display() # Expected: 0 -> 1 -> 3

my_list.delete_node(5) # Try to delete a non-existent node
print("After trying to delete 5:") # Expected: "Node with data '5' not found in the list."
my_list.display() # Expected: 0 -> 1 -> 3 (list remains unchanged)

my_list.delete_node(3) # Delete the last node
print("After deleting 3:")
my_list.display() # Expected: 0 -> 1

my_list.delete_node(0) # Delete the new head
print("After deleting 0:")
my_list.display() # Expected: 1

my_list.delete_node(1) # Delete the last remaining node
print("After deleting 1:")
my_list.display() # Expected: (empty line)

my_list.delete_node(100) # Delete from an empty list
print("After deleting from empty:") # Expected: "Node with data '100' not found in the list."
my_list.display() # Expected: (empty line)

print("\n--- Testing Complete ---")
```

```python
# Define a Node class for the Doubly Linked List
class DoublyNode:
    """
    Represents a single node in a doubly linked list.
    """
    def __init__(self, data):
        # The data stored in the node
        self.data = data
        # A pointer to the next node in the list, initially None
        self.next = None
        # A pointer to the previous node in the list, initially None
        self.prev = None

# Define the DoublyLinkedList class
class DoublyLinkedList:
    """
    Represents a doubly linked list.
    """
    def __init__(self):
        # The head of the list, initially None
        self.head = None
        # The tail of the list, initially None (useful for O(1) append)
        self.tail = None

    def append(self, data):
        """
        Adds a new node with the given data to the end of the list.
        This is an O(1) operation with a tail pointer.
        """
        new_node = DoublyNode(data)

        # If the list is empty
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            return

        # If the list is not empty, link the current tail to the new node
        self.tail.next = new_node
        new_node.prev = self.tail # Link the new node back to the old tail
        self.tail = new_node # Update the tail to the new node

    def prepend(self, data):
        """
        Adds a new node with the given data to the beginning of the list.
        This is an O(1) operation.
        """
        new_node = DoublyNode(data)

        # If the list is empty
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            return

        # If the list is not empty, link the new node to the current head
        new_node.next = self.head
        self.head.prev = new_node # Link the current head back to the new node
        self.head = new_node # Update the head to the new node

    def delete_node(self, key):
        """
        Deletes the first node found with the given data (key).
        This is an O(n) operation in the worst case.
        Handles deleting head, tail, and middle nodes.
        """
        current_node = self.head

        # Traverse the list to find the node with the key
        while current_node is not None and current_node.data != key:
            current_node = current_node.next

        # If the node was not found
        if current_node is None:
            print(f"Node with data '{key}' not found in the list.")
            return

        # Case 1: The node to be deleted is the head
        if current_node == self.head:
            self.head = current_node.next
            if self.head: # If the list is not empty after deleting the head
                self.head.prev = None # The new head has no previous node
            else: # If the list becomes empty
                self.tail = None
            current_node.next = None # Dereference the deleted node
            # current_node.prev is already None
            del current_node
            return

        # Case 2: The node to be deleted is the tail
        if current_node == self.tail:
            self.tail = current_node.prev
            self.tail.next = None # The new tail has no next node
            current_node.prev = None # Dereference the deleted node
            # current_node.next is already None
            del current_node
            return

        # Case 3: The node to be deleted is in the middle
        # Link the previous node to the next node
        current_node.prev.next = current_node.next
        # Link the next node back to the previous node
        current_node.next.prev = current_node.prev

        # Dereference the deleted node
        current_node.next = None
        current_node.prev = None
        del current_node


    def display_forward(self):
        """
        Prints the data of each node from head to tail.
        """
        current_node = self.head
        elements = []
        while current_node:
            elements.append(str(current_node.data))
            current_node = current_node.next
        print(" -> ".join(elements))

    def display_backward(self):
        """
        Prints the data of each node from tail to head.
        """
        current_node = self.tail
        elements = []
        while current_node:
            elements.append(str(current_node.data))
            current_node = current_node.prev
        print(" <- ".join(elements))

    def search(self, key):
        """
        Searches for a node with the given data (key).
        Can search from head or tail. O(n) in worst case.
        """
        current_node = self.head
        while current_node:
            if current_node.data == key:
                return True # Found the key
            current_node = current_node.next
        return False # Key not found


# --- Simple Test Case ---

print("--- Testing Doubly Linked List ---")

# Create an empty doubly linked list
my_doubly_list = DoublyLinkedList()
print("\nInitial list (forward):")
my_doubly_list.display_forward() # Expected: (empty line)
print("Initial list (backward):")
my_doubly_list.display_backward() # Expected: (empty line)


# Test append operation
print("\nAppending nodes:")
my_doubly_list.append(1)
my_doubly_list.append(2)
my_doubly_list.append(3)
print("List after appends (forward):")
my_doubly_list.display_forward() # Expected: 1 -> 2 -> 3
print("List after appends (backward):")
my_doubly_list.display_backward() # Expected: 3 <- 2 <- 1

# Test prepend operation
print("\nPrepending nodes:")
my_doubly_list.prepend(0)
my_doubly_list.prepend(-1)
print("List after prepends (forward):")
my_doubly_list.display_forward() # Expected: -1 -> 0 -> 1 -> 2 -> 3
print("List after prepends (backward):")
my_doubly_list.display_backward() # Expected: 3 <- 2 <- 1 <- 0 <- -1

# Test search operation
print("\nTesting search:")
print(f"Is 2 in the list? {my_doubly_list.search(2)}")   # Expected: True
print(f"Is 5 in the list? {my_doubly_list.search(5)}")   # Expected: False
print(f"Is -1 in the list? {my_doubly_list.search(-1)}") # Expected: True

# Test delete operation
print("\nDeleting nodes:")
my_doubly_list.delete_node(2) # Delete node with data 2 (middle)
print("After deleting 2 (forward):")
my_doubly_list.display_forward() # Expected: -1 -> 0 -> 1 -> 3
print("After deleting 2 (backward):")
my_doubly_list.display_backward() # Expected: 3 <- 1 <- 0 <- -1

my_doubly_list.delete_node(-1) # Delete the head node
print("After deleting -1 (forward):")
my_doubly_list.display_forward() # Expected: 0 -> 1 -> 3
print("After deleting -1 (backward):")
my_doubly_list.display_backward() # Expected: 3 <- 1 <- 0

my_doubly_list.delete_node(3) # Delete the tail node
print("After deleting 3 (forward):")
my_doubly_list.display_forward() # Expected: 0 -> 1
print("After deleting 3 (backward):")
my_doubly_list.display_backward() # Expected: 1 <- 0

my_doubly_list.delete_node(5) # Try to delete a non-existent node
print("After trying to delete 5 (forward):") # Expected: "Node with data '5' not found in the list."
my_doubly_list.display_forward() # Expected: 0 -> 1 (list remains unchanged)
print("After trying to delete 5 (backward):")
my_doubly_list.display_backward() # Expected: 1 <- 0

my_doubly_list.delete_node(0) # Delete the current head
print("After deleting 0 (forward):")
my_doubly_list.display_forward() # Expected: 1
print("After deleting 0 (backward):")
my_doubly_list.display_backward() # Expected: 1

my_doubly_list.delete_node(1) # Delete the last remaining node
print("After deleting 1 (forward):")
my_doubly_list.display_forward() # Expected: (empty line)
print("After deleting 1 (backward):")
my_doubly_list.display_backward() # Expected: (empty line)


print("\n--- Testing Complete ---")
```

##Graph Algorithm

```python
from collections import deque

# Define the graph
graph = {}
graph["you"] = ["alice", "bob", "claire"]
graph["bob"] = ["anuj", "peggy"]
graph["alice"] = ["peggy"]
graph["claire"] = ["thom", "jonny"]
graph["anuj"] = []
graph["peggy"] = []
graph["thom"] = []
graph["jonny"] = []

# Function to check if a person is a seller (ends with 'm')
def person_is_seller(name):
  """Checks if a person's name ends with 'm'."""
  return name[-1] == 'm'

# Breadth-First Search function
def search(name):
  """
  Performs a BFS starting from 'name' to find a seller.

  Args:
    name: The starting person's name (key in the graph).

  Returns:
    True if a seller is found, False otherwise.
  """
  # Create a new queue (Double-ended queue for efficient appending and popping)
  search_queue = deque()
  # Add all of the starting person's neighbors to the queue
  search_queue.extend(graph[name]) # Use extend for adding multiple items

  # Keep track of people already searched to avoid infinite loops
  searched = set()

  # While the queue is not empty
  while search_queue:
    # Grab the first person off the queue
    person = search_queue.popleft()

    # Only search this person if you haven't already
    if person not in searched:
      # Check if the person is a seller
      if person_is_seller(person):
        print(person + " is a mango seller!")
        return True # Found the seller, exit the function and return True
      else:
        # If not a seller, add all of their friends to the queue
        # Use extend to add multiple items efficiently
        search_queue.extend(graph[person])
        # Add the person to the set of searched people
        searched.add(person)

  # If the loop finishes and the queue is empty, the seller was not found
  return False

# Start the search from "you"
print(search("you")) # This will print the result of the search
```

## Trees

```python
import os
from collections import deque

def traverse_directory_bfs(start_dir):
  """
  Traverses a directory structure using Breadth-First Search (BFS).

  Prints the full path of each file encountered.
  Adds subdirectories to a queue to be visited later.

  Args:
    start_dir: The path to the starting directory for traversal.
  """
  # Use a deque as a queue for efficient appending and popping
  dirs_to_visit = deque()

  # Add the starting directory to the queue
  dirs_to_visit.append(start_dir)

  print(f"Starting BFS traversal from: {start_dir}\n")

  # While there are directories left to visit in the queue
  while dirs_to_visit:
    # Get the next directory path from the front of the queue
    current_dir = dirs_to_visit.popleft()

    try:
      # List all entries (files and subdirectories) in the current directory
      with os.scandir(current_dir) as entries:
        for entry in entries:
          # Construct the full path for the entry
          full_path = os.path.join(current_dir, entry.name)

          # Check if the entry is a file
          if entry.is_file():
            # If it's a file, print its name (or full path)
            print(f"File: {full_path}")

          # Check if the entry is a directory
          elif entry.is_dir():
            # If it's a directory, add it to the end of the queue to visit later
            dirs_to_visit.append(full_path)

    except FileNotFoundError:
      print(f"Error: Directory not found - {current_dir}")
    except PermissionError:
      print(f"Error: Permission denied to access - {current_dir}")
    except Exception as e:
      print(f"An unexpected error occurred processing {current_dir}: {e}")

  print("\nBFS traversal finished.")
```

## Dijkstra's Algorithm

```python
import heapq
from collections import defaultdict

def dijkstra(graph, start_node):
    """
    Finds the shortest paths from a start_node to all other nodes
    in a graph with non-negative edge weights using Dijkstra's algorithm.

    Args:
        graph: A dictionary representing the graph.
               Keys are nodes, values are lists of (neighbor, weight) tuples.
               Example: {'A': [('B', 1), ('C', 4)], 'B': [('C', 2)]}
        start_node: The node to start the shortest path search from.

    Returns:
        A dictionary where keys are nodes and values are the shortest
        distance from the start_node to that node.
    """

    # 1. Initialize distances:
    # Create a dictionary to store the shortest distance found so far for each node.
    # Initialize distance to start_node as 0 and all other nodes as infinity.
    # We use float('inf') to represent infinity.
    distances = defaultdict(lambda: float('inf'))
    distances[start_node] = 0

    # 2. Initialize priority queue:
    # A priority queue (min-heap) is used to efficiently extract the node
    # with the smallest distance found so far.
    # We store tuples of (distance, node). The distance is the priority.
    priority_queue = [(0, start_node)] # Start with the start_node and its distance (0)

    # 3. Initialize visited set:
    #    Keep track of nodes for which we have finalized the shortest path.
    visited = set()

    # 4. Process nodes while the priority queue is not empty:
    while priority_queue:
        # a. Extract the node with the smallest distance from the priority queue:
        #    heapq.heappop extracts the smallest item (based on the first element of the tuple).
        current_distance, current_node = heapq.heappop(priority_queue)

        # b. If the node has already been visited, skip it:
        #    We might encounter the same node multiple times in the queue if we found
        #    a shorter path to it after it was initially added. We only process it
        #    the first time it's extracted with the truly shortest distance.
        if current_node in visited:
            continue

        # c. Mark the current node as visited:
        #    We have now found the shortest path to this node.
        visited.add(current_node)

        # d. Explore neighbors:
        #    Check all adjacent nodes (neighbors) of the current_node.
        #    graph.get(current_node, []) handles cases where a node might not be a key
        #    in the graph dictionary (e.g., a destination with no outgoing edges).
        for neighbor, weight in graph.get(current_node, []):
            # i. Calculate the distance to the neighbor through the current_node:
            #    This is the distance to the current_node plus the weight of the edge
            #    from current_node to neighbor.
            distance_through_current = current_distance + weight

            # ii. If this path is shorter than the best distance found so far for the neighbor:
            if distance_through_current < distances[neighbor]:
                # Update the shortest distance to the neighbor
                distances[neighbor] = distance_through_current

                # Push the neighbor onto the priority queue with its new, shorter distance
                # This ensures it will be processed later when it has the smallest distance
                heapq.heappush(priority_queue, (distance_through_current, neighbor))

    # 5. Return the distances:
    #    Once the priority queue is empty, we have found the shortest paths to all reachable nodes.
    return distances

# ---Example ---

# Represent the graph using an adjacency list
# 'A': [('B', 1), ('C', 4)] means from node A, there's an edge to B with weight 1,
# and an edge to C with weight 4.
example_graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('C', 2), ('D', 5)],
    'C': [('D', 1)],
    'D': [('E', 3)],
    'E': [] # Node E has no outgoing edges
}

# Specify the starting node
start = 'A'

# Run Dijkstra's algorithm
shortest_distances = dijkstra(example_graph, start)

# Print the results
print(f"Shortest distances from node {start}:")
for node, distance in sorted(shortest_distances.items()):
     # Check if distance is still infinity, meaning the node is unreachable
     if distance == float('inf'):
         print(f"  To {node}: Unreachable")
     else:
         print(f"  To {node}: {distance}")

# Expected Output:
# Shortest distances from node A:
#   To A: 0
#   To B: 1
#   To C: 3  (A -> B -> C)
#   To D: 4  (A -> B -> C -> D)
#   To E: 7  (A -> B -> C -> D -> E)
```

## Greedy Algorithm

```python
def greedy_set_cover(states_needed, stations):
    """
    Finds a near-optimal set of radio stations to cover a set of states
    using a greedy approach.

    Args:
        states_needed: A set of strings representing the states that need coverage.
        stations: A dictionary where keys are station names (strings) and values
                  are sets of strings representing the states each station covers.

    Returns:
        A set of station names that provides coverage for all states_needed.
        This is a near-optimal solution found by the greedy algorithm.
    """

    # Initialize the set of states that still need to be covered.
    # We make a copy of the initial set so we don't modify the original input.
    states_to_cover = states_needed.copy()

    # Initialize the set of stations we've chosen for the coverage.
    final_stations = set()

    # Loop while there are still states that need coverage.
    # The loop continues as long as the 'states_to_cover' set is not empty.
    while states_to_cover:
        # In each iteration, we want to find the station that covers the most
        # *remaining* uncovered states.

        # Initialize variables to keep track of the best station found in this iteration
        # and the set of states it covers that are currently needed.
        best_station = None
        states_covered_by_best_station = set()

        # Iterate through each available station and the states it covers.
        for station, states_for_station in stations.items():
            # Use set intersection to find which states this station covers
            # that are still in the 'states_to_cover' set.
            # This 'covered' set represents the *new* states this station would cover.
            covered = states_to_cover.intersection(states_for_station)
            # Equivalently: covered = states_to_cover & states_for_station

            # Check if this station covers more of the *remaining* states
            # than the best station found so far in this iteration.
            if len(covered) > len(states_covered_by_best_station):
                # If it does, update our 'best_station' and 'states_covered_by_best_station'.
                best_station = station
                states_covered_by_best_station = covered

        # After checking all stations, we have found the one that covers
        # the largest number of currently uncovered states.

        # If no station covers any *new* states (e.g., all remaining states
        # are only covered by stations already chosen or are unreachable),
        # the 'best_station' will still be None. We should break the loop
        # if that happens to avoid infinite loops if not all states are coverable.
        # (Though in a typical Set Cover problem formulation, all elements are assumed coverable)
        if best_station is None:
             print("Warning: Not all states could be covered by the available stations.")
             break # Exit the loop if no progress can be made

        # Add the chosen 'best_station' to our set of final stations.
        final_stations.add(best_station)

        # Remove the states covered by the 'best_station' from the set of
        # states that still need coverage.
        # We use set difference to efficiently remove the covered states.
        states_to_cover -= states_covered_by_best_station
        # Equivalently: states_to_cover = states_to_cover.difference(states_covered_by_best_station)

        # Optional: Print progress
        # print(f"Chosen station: {best_station}. States covered in this step: {states_covered_by_best_station}")
        # print(f"Remaining states to cover: {states_to_cover}")


    # The loop finishes when 'states_to_cover' is empty.
    # Return the set of chosen stations.
    return final_stations



# 1. Define the set of all states we need to cover.
all_states = set(["mt", "wa", "or", "id", "nv", "ut", "ca"])

# 2. Define the available radio stations and the states each covers.
radio_stations = {
    "kone": {"id", "nv", "ut"},
    "ktwo": {"wa", "id", "mt"},
    "kthree": {"or", "nv", "ca"},
    "kfour": {"nv", "ut"},
    "kfive": {"ca", "az"}
}

# 3. Run the greedy algorithm
chosen_stations = greedy_set_cover(all_states, radio_stations)

# 4. Print the result
print("\nStates to cover:", all_states)
print("Available stations:", radio_stations)
print("\nGreedy algorithm suggests these stations:", chosen_stations)

# --- Another Example ---
print("-" * 30)
all_states_2 = set(["a", "b", "c", "d", "e", "f"])
radio_stations_2 = {
    "s1": {"a", "b", "c"},
    "s2": {"a", "d"},
    "s3": {"e", "f"},
    "s4": {"b", "c", "d"},
    "s5": {"e", "a"}
}

chosen_stations_2 = greedy_set_cover(all_states_2, radio_stations_2)
print("\nStates to cover:", all_states_2)
print("Available stations:", radio_stations_2)
print("\nGreedy algorithm suggests these stations:", chosen_stations_2)
# Expected for second example (greedy might pick s1, then s3, then s2/s5 for 'a'):
# s1: {a, b, c} -> remaining {d, e, f}
# s3: {e, f}   -> remaining {d}
# s2: {a, d}   -> covers {d}
# Result: {s1, s3, s2} - covers all.
```
